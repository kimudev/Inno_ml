**Задача 1.** Создать матрицу размером 8х8 элементов, состоящую из нулей. Заполнить эту матрицу значениями,расположенными в шахматном порядке. Задачу решить через срезы массива NumPy.

```python
import numpy as np

matrix = np.zeros((8, 8), dtype=int)

matrix[::2, 1::2] = 1  # Заполняем 1 в нечетных колонках для четных строк
matrix[1::2, ::2] = 1  # Заполняем 1 в четных колонках для нечетных строк

print(matrix)
```

**Задача 2.** Средствами NumPy сформируйте вектор из 2000 случайных величин, распределенных по нормальному закону с математически можиданием 0 и дисперсией 5. Средствами NumPy вычислите для полученного вектора экспериментальное значение математического ожидания(среднее значение) и дисперсию.

```python
mean = 0  # математическое ожидание
variance = 5  # дисперсия

A = np.random.normal(loc=mean, scale=np.sqrt(variance), size=2000)

experimental_mean = np.mean(A)
experimental_variance = np.var(A)

print(f'Математическое ожидание: {experimental_mean}\nДисперсия: {experimental_variance}')
```
